Problem:

$$ -\epsilon \nabla^2u + b\nabla u = 0,\;\;\textrm{in} \;\;\Omega = [0,1]\times[0,1]$$

$$ u = 0\;\;\;\forall x=0,x=1 $$

$$ u = 1\;\;\;\forall y=0,y=1 $$

Variational formulation:

$$ a(u,v) = l(v)$$

$$ a(u,v) := \epsilon \iint_\Omega \nabla u \nabla v dx\,dy +b \iint_\Omega \nabla u \;v\,dx\,dy$$

$$ l(v) := 0 $$

```python
from firedrake import *
import matplotlib.pyplot as plt

mesh = UnitSquareMesh(50,50)
plot(mesh)

e = Constant(1) # diffusion constant 
b = Constant(1) # convection constant

V = FunctionSpace(mesh, 'CG', 1)
u = TrialFunction(V)
v = TestFunction(V)
x,y = SpatialCoordinate(mesh)
f = Function(V)
f = Constant(0)


boundary_ids1 = {1,2} # boundaries x=0 and x=1
boundary_ids2 = {3,4} # boundaries y=0 and y=1

bc1 = DirichletBC(V,0,boundary_ids1) # Boundary conditions for x=0 and x=1
bc2 = DirichletBC(V,1,boundary_ids2) # Boundary conditions for y=0 and y=1

uh = Function(V)
L = f*v*dx
a = ( e * dot(grad(v),grad(u)) ) * dx + ( b * u.dx(0) * v * dx)

solve(a == L, uh, bcs=[bc1,bc2], solver_parameters={"ksp_type": "cg"})
plot(uh)

plt.rcParams['figure.figsize']=(15,14)
```

```python
u_exact = ( exp(b*x/e) - 1 )*( exp(b*y/e) - 1 ) / ( exp(b/e) - 1 )**2
difference = assemble(interpolate(u_exact, V) - uh)
plot(difference)
```

# 1-d

Problem:

$$ -\epsilon u'' + bu' = 0,\;\;\textrm{in} \;\;\Omega = [0,1]$$

$$ u = 0,\;\;\; x=0 $$

$$ u = 1,\;\;\; x=1 $$

Variational formulation:

$$ a(u,v) = l(v)$$

$$ a(u,v) := \epsilon \int_\Omega u'v' dx\, +b \int_\Omega u'v\,dx$$

$$ l(v) := 0 $$

```python
from firedrake import *
import matplotlib.pyplot as plt

N = 10
mesh = UnitIntervalMesh(N)

e = 1  # diffusion constant 
b = 200  # convection constant

V = FunctionSpace(mesh, 'CG', 1)
u = TrialFunction(V)
v = TestFunction(V)


x = SpatialCoordinate(mesh)[0]

f = Function(V)
f = Constant(0)


boundary_ids1 = {1} # boundaries x=0 
boundary_ids2 = {2} # boundaries x=1 

bc1 = DirichletBC(V,0,boundary_ids1) # Boundary conditions for x=0
bc2 = DirichletBC(V,1,boundary_ids2) # Boundary conditions for x=1

uh = Function(V)
L = f*v*dx
a = ( e * dot(grad(v),grad(u)) ) * dx + ( b * u.dx(0) * v * dx)

solve(a == L, uh, bcs=[bc1,bc2])
pe = b/(2*e*N)
print ('Pe =',pe)
plot(uh)
plt.title(pe) 


u_exact = (exp(b*x/e)-1)/(exp(b/e)-1)
difference = assemble(interpolate(u_exact, V) - uh)
plot(difference)
plt.title('Error')
plt.show()
```

```python
u_exact = (exp(b*x/e)-1)/(exp(b/e)-1)
difference = assemble(interpolate(u_exact, V) - uh)
plot(difference)
```

```python
from firedrake import *
import matplotlib.pyplot as plt

N = 10
mesh = UnitIntervalMesh(N)

e = 1  # diffusion constant 
b = 2  # convection constant

V = FunctionSpace(mesh, 'CG', 1)
u = TrialFunction(V)
v = TestFunction(V)


x = SpatialCoordinate(mesh)[0]

f = Function(V)
f = 10*exp(-5*x) - 4*exp(-x)


boundary_ids1 = {1} # boundaries x=0 
boundary_ids2 = {2} # boundaries x=1 

bc1 = DirichletBC(V,0,boundary_ids1) # Boundary conditions for x=0
bc2 = DirichletBC(V,1,boundary_ids2) # Boundary conditions for x=1

uh = Function(V)
L = f*v*dx
a = ( e * dot(grad(v),grad(u)) ) * dx + ( b * u.dx(0) * v * dx)

solve(a == L, uh, bcs=[bc1,bc2])
pe = b/(2*e*N)
print ('Pe =',pe)
plot(uh)
plt.title(pe) 
plt.show()
```
