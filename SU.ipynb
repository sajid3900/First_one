# Streamline-Upwind

Consiider the convection-diffusion equation : 

$$ -\epsilon u'' + bu' = 0,\;\;\textrm{in} \;\;\Omega = [0,1]$$

$$ u = 0,\;\;\; x=0 $$

$$ u = 1,\;\;\; x=1 $$

where, $\epsilon$ and b are diffusion and convection coefficients respectively.

Variational formulation:

$$ a(u,v) = l(v)$$

$$ a(u,v) := \epsilon \int_\Omega u'v' dx\, +b \int_\Omega u'v\,dx$$

$$ l(v) := 0 $$

Solving it by galerkin method gives oscillatory results for Peclet number Pe>1.
One of the ways of stabilizing is by adding an extra diffusion coeffient.

For stabilization, an artificial diffusion coefficient $\epsilon_h$ is added to the diffusion coefficient. Now the bilinear form is defined as folows.



$$ a(u,v) := (\epsilon + \epsilon_h) \int_\Omega u'v' dx\, +b \int_\Omega u'v\,dx$$

The value of $\epsilon_h$ is chosen such that it reduces the Peclet number. Choose $ \epsilon_h = \frac{\gamma}{2} hb$. 

If $\gamma \geq 1$, the effective Peclet number is sufficiently small and hence diffusion is minimized.

For perfect stabilization use $\gamma = \textrm{coth}(Pe)-1/Pe$

```python
from firedrake import *
import matplotlib.pyplot as plt
import numpy as np

N = 10
mesh = UnitIntervalMesh(N)

e = 1  # diffusion constant 
b = 500 # convection constant
h = 1/N
pe = b*h/(2*e) # Peclet Number
#gamma = 1
gamma = ( ((exp(2*pe)+1)/(exp(2*pe)-1)) - 1/pe)  # For perfect stabilization
eh = gamma/2*h*b # artificial diffusion coefficient
pe_eff = b*h/(2*(e+eh)) # Effective Peclet number

V = FunctionSpace(mesh, 'CG', 1)
u = TrialFunction(V)
v = TestFunction(V)

x = SpatialCoordinate(mesh)[0]

f = Function(V)
f = Constant(0)


boundary_ids1 = {1} # boundaries x=0 
boundary_ids2 = {2} # boundaries x=1 

bc1 = DirichletBC(V,0,boundary_ids1) # Boundary conditions for x=0, u=0
bc2 = DirichletBC(V,1,boundary_ids2) # Boundary conditions for x=1, u=1

uh = Function(V)
L = f*v*dx
a = ( (e+eh) * dot(grad(v),grad(u)) ) * dx + b * u.dx(0) * v * dx 

solve(a == L, uh, bcs=[bc1,bc2])

print ('Pe =',pe)
print ('Effective Pe =',pe_eff)
plot(uh, interactive = 'TRUE')
plt.title(pe) 

File("su.pvd").write(uh)

u_exact = (exp(b*x/e)-1)/(exp(b/e)-1)
difference = assemble(interpolate(u_exact, V) - uh)
plot(difference)
plt.title('Error') 
plt.show()
```

# 2-d

```python
from firedrake import *
import matplotlib.pyplot as plt


N = 10
mesh = UnitSquareMesh(N,N)
#plot(mesh)
h = 1/N

e = 1  # diffusion constant 
b = 10  # convection constant

V = FunctionSpace(mesh, 'CG', 1)
u = TrialFunction(V)
v = TestFunction(V)
x,y = SpatialCoordinate(mesh)
f = Function(V)
f = Constant(0)


boundary_ids1 = {1,2} # boundaries x=0 and x=1
boundary_ids2 = {3,4} # boundaries y=0 and y=1

bc1 = DirichletBC(V,0,boundary_ids1) # Boundary conditions for x=0 and x=1
bc2 = DirichletBC(V,1,boundary_ids2) # Boundary conditions for y=0 and y=1

pe = b/(2*e*N)

uh = Function(V)
L = f*v*dx
a = ( e * dot(grad(v),grad(u)) ) * dx + ( b * u.dx(0) * v * dx) + (b*h/2) * dot(grad(v),grad(u))  * dx

solve(a == L, uh, bcs=[bc1,bc2])

print ('Pe =',pe)
plot(uh, plot3d = 'TRUE')
plt.rcParams['figure.figsize']=(15,14)
plt.title(pe) 


```

```python
u_exact = ( exp(b*x/e) - 1 )*( exp(b*y/e) - 1 ) / ( exp(b/e) - 1 )**2
difference = assemble(interpolate(u_exact, V) - uh)
plot(difference)
```
