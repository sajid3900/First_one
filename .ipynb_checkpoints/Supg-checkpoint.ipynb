# SUPG 


Consiider the convection-diffusion equation : 

$$ -\epsilon u'' + bu' = 0,\;\;\textrm{in} \;\;\Omega = [0,1]$$

$$ u = 0,\;\;\; x=0 $$

$$ u = 1,\;\;\; x=1 $$

where, $\epsilon$ and b are diffusion and convection coefficients respectively.

In SUPG method, the modified test function used
$$v + \delta b v'$$

Multiplying by test function $ v+\delta b v' $ on both sides and integrating

$$ \int_\Omega (-\epsilon u''+bu')(v + \delta b v') dx = 0 $$

$$ \epsilon \int_\Omega u'v'dx + b\int_\Omega u'vdx + \delta b^2 \int_\Omega u'v'dx = 0 $$  

Variational formulation:

$$ a(u,v) = l(v)$$

$$ a(u,v) := \epsilon \int_\Omega u'v'dx + b\int_\Omega u'vdx + \delta b^2 \int_\Omega u'v'dx $$

$$ l(v) := 0 $$

```python
from firedrake import *
import matplotlib.pyplot as plt
#from sympy.mpmath import *

N = 10
mesh = UnitIntervalMesh(N)

e = 1  # diffusion constant 
b = 200 # convection constant

V = FunctionSpace(mesh, 'CG', 1)
u = TrialFunction(V)
v = TestFunction(V)


x = SpatialCoordinate(mesh)[0]

f = Function(V)
f = Constant(0)


boundary_ids1 = {1} # boundaries x=0 
boundary_ids2 = {2} # boundaries x=1 

bc1 = DirichletBC(V,0,boundary_ids1) # Boundary conditions for x=0
bc2 = DirichletBC(V,1,boundary_ids2) # Boundary conditions for x=1

h = 1/N
pe = b/(2*e*N)
#dell =0.001 
dell = (h/(2*b))*( ((exp(2*pe)+1)/(exp(2*pe)-1)) - 1/pe)

uh = Function(V)
L = f*v*dx
a = ( e * dot(grad(v),grad(u)) ) * dx + ( b * u.dx(0) * v * dx) + dell * b * b * dot(grad(v),grad(u)) * dx

solve(a == L, uh, bcs=[bc1,bc2])

print ('Pe =',pe)
plot(uh)
plt.title(pe) 

u_exact = Function(V)
u_exact = (exp(b*x/e)-1)/(exp(b/e)-1)
difference = assemble(interpolate(u_exact, V) - uh)
plot(difference)
plt.title('Error') 
plt.show()
```

```python
from firedrake import *
import matplotlib.pyplot as plt


N = 10
mesh = UnitSquareMesh(N,N)
#plot(mesh)
h = 1/N

e = 1  # diffusion constant 
b = 10  # convection constant

V = FunctionSpace(mesh, 'CG', 1)
u = TrialFunction(V)
v = TestFunction(V)
x,y = SpatialCoordinate(mesh)
f = Function(V)
f = Constant(0)


boundary_ids1 = {1,2} # boundaries x=0 and x=1
boundary_ids2 = {3,4} # boundaries y=0 and y=1

bc1 = DirichletBC(V,0,boundary_ids1) # Boundary conditions for x=0 and x=1
bc2 = DirichletBC(V,1,boundary_ids2) # Boundary conditions for y=0 and y=1

pe = b/(2*e*N)
dell = (h/(2*b))*( ((exp(2*pe)+1)/(exp(2*pe)-1)) - 1/pe)

uh = Function(V)
L = f*v*dx
a = ( e * dot(grad(v),grad(u)) ) * dx + ( b * u.dx(0) * v * dx) + dell * dot(b*grad(v),b*grad(u)) * dx

solve(a == L, uh, bcs=[bc1,bc2])

print ('Pe =',pe)
plot(uh,plot3d = 'TRUE')
plt.rcParams['figure.figsize']=(15,14)
plt.title(pe) 

```

```python
u_exact = ( exp(b*x/e) - 1 )*( exp(b*y/e) - 1 ) / ( exp(b/e) - 1 )**2
difference = assemble(interpolate(u_exact, V) - uh)
plot(difference)
```
